```python
import numpy as np
import pickle
import matplotlib.pyplot as plt
import matplotlib as mpl

mpl.rcParams["lines.linewidth"] = 2.5
mpl.rcParams["savefig.bbox"] = "tight"
mpl.rcParams["figure.figsize"] = (8,5)
mpl.rcParams["font.size"] = 13
# mpl.rcParams["legend.frameon"] = False
mpl.rcParams["legend.frameon"] = True
mpl.rcParams["legend.framealpha"] = 1
```

# Different $\psi_1$ values:

```python
xi_g = 1000.
xi_p = xi_g
xi_a = 1000.

psi_0 = 0.003


NUM_FIG = 7
fig, axes = plt.subplots(NUM_FIG, 1, figsize=(8, NUM_FIG * 5))


for psi_1 in [0.5, 0.7, 0.9]:
    res = pickle.load(open("simul_data/simul_psi_0_{:.3f}_psi_1_{:.3f}_xi_a_{}_xi_g_{}".format(psi_0, psi_1, xi_a, xi_g), "rb"))
    axes[0].plot(res["x"]*100 / 0.115**2, label="ψ_0 = {:.3f}, ψ_1 = {:.1f}".format(psi_0, psi_1))
    axes[1].plot(res["i"], label="ψ_0 = {:.3f}, ψ_1 = {:.1f}".format(psi_0, psi_1))
    axes[2].plot(res["e"], label="ψ_0 = {:.3f}, ψ_1 = {:.1f}".format(psi_0, psi_1))
    axes[3].plot(res["states"][:, 1], label="ψ_0 = {:.3f}, ψ_1 = {:.1f}".format(psi_0, psi_1))
    axes[4].plot(np.exp(res["states"][:, 2]), label="ψ_0 = {:.3f}, ψ_1 = {:.1f}".format(psi_0, psi_1))
    axes[5].plot(res["distorted_tech_prob"], label="ψ_0 = {:.3f}, ψ_1 = {:.1f}".format(psi_0, psi_1))
#     axes[6].plot(res["distorted_tech_prob"], label="ψ_0 = {:.3f}, ψ_1 = {:.1f}".format(psi_0, psi_1))
    axes[6].plot(res["distorted_damage_prob"], label="ψ_0 = {:.3f}, ψ_1 = {:.1f}".format(psi_0, psi_1))


axes[0].set_ylabel("% of GDP")
axes[0].set_title("R & D investment as percentage of GDP")
axes[1].set_title("Physical investment")
axes[2].set_title("Emission")
axes[3].set_title("Temperature anomaly")
axes[4].set_title("I_g, technology jump intensity")
axes[5].set_title("technology probability")
# axes[6].set_title("second technology jump probability\n assuem first jump happens at year 40")
axes[6].set_title("Damage jump probability")


for ax in axes:
    ax.legend()
    ax.set_xlabel("Years")
    ax.set_ylim(0)
    ax.grid()

fig.get_axes()[0].annotate("Fix ψ_0 = {:.3f}, compare ψ_1".format(psi_0), (0.5, 0.943),
                              xycoords="figure fraction", ha="center",
                           fontsize=18
                          )
fig.tight_layout()

plt.savefig("figures/psi_0_{:.3f}.pdf".format(psi_0))
```
